#### How are in-demand skills trending for Data Analysts?
Methology
1.	Aggregate skill counts on a monthly basis
2.	Re-analyze based on percentage of total jobs
3.	Plot the monthly skill demand

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import seaborn as sns
import matplotlib.pyplot as plt  
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")].copy()
```

```python
# Create a new column for month number
df_DA_US["job_posted_month_no"] = df_DA_US["job_posted_date"].dt.month

# Explode job_skills
df_DA_US_explode = df_DA_US.explode("job_skills")

# Pivot the exploded df
df_DA_US_pivot = df_DA_US_explode.pivot_table(index="job_posted_month_no", columns="job_skills",\
aggfunc="size", fill_value=0)

# Create a row "Total"
df_DA_US_pivot.loc["Total"] = df_DA_US_pivot.sum()

# Sort row "Total"
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc["Total"].sort_values(ascending=False).index]

# Drop the row "Total"
df_DA_US_pivot = df_DA_US_pivot.drop("Total")
```

```python
# Plot top 5 skills for data analysts, indexed by month
df_DA_US_pivot.iloc[:, :5].plot(kind="line")
plt.title("Trending Top Skills for Data Analysts in the US")
plt.xlabel("Months")
plt.ylabel("Count")
plt.show()
```

```python
# Get monthly totals
DA_totals = df_DA_US.groupby("job_posted_month_no").size()
DA_totals
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals, axis=0)*100

# Turn month numbers into month names

# turn job_posted_month_no into a column
df_DA_US_percent = df_DA_US_percent.reset_index() 
# create a column with month names
df_DA_US_percent["job_posted_month"] = df_DA_US_percent["job_posted_month_no"]\
    .apply(lambda x: pd.to_datetime(x, format="%m").strftime("%b"))
# set column with month names as an index
df_DA_US_percent = df_DA_US_percent.set_index("job_posted_month")
# drop month_no column
df_DA_US_percent = df_DA_US_percent.drop(columns="job_posted_month_no")

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, palette="tab10")
sns.set_theme(style="ticks")
sns.despine()

plt.title("Trending Top Skills for Data Analysts in the US")
plt.ylabel("Likelihood in Job Posting")
plt.xlabel("2023")
plt.legend().remove()

# annotate the plot with the top 5 skills using plt.text()
from matplotlib.ticker import PercentFormatter
ax = plt.gca()
lines = ax.get_lines()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(df_plot.shape[1]): # gives number of columns
    line = lines[i]
    color = line.get_color()
    y = df_plot.iloc[2, i]  # 3rd row's value of column i
    label = df_plot.columns[i]
    plt.text(x=len(df_plot)-0.5, y=y, s=label, color=color)

plt.show()
# Alternative option
# for i in range(5):
#     plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
