```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df["job_posted_month"] = df["job_posted_date"].dt.strftime("%b")
```

```python
# make data frame for each month using a loop
months = df["job_posted_month"].unique()

# write this with dict comprehension

dict_months = {month: df[df["job_posted_month"] == month] for month in months}
```

```python
# Create a list of DataFrames for each month in Q1
quarter_1 = [dict_months["Jan"], dict_months["Feb"], dict_months["Mar"]]
# quarter_1 = {month: dict_months[month] for month in ["Jan", "Feb", "Mar"]} # <--- This 
# is the dict comprehension version

# Concatenate the DataFrames in the list into a single DataFrame
df_q1 = pd.concat(quarter_1, ignore_index=True)

# Display the concatenated DataFrame
df_q1  
```

```python
# Filter df for Data Analyst roles
df_DA = df[(df['job_title_short'] == 'Data Analyst')].copy()

# Create a column for month with month name consisting of the first three letters
df_DA["job_posted_month"] = df_DA["job_posted_date"].dt.strftime("%b")

months = df_DA['job_posted_month'].unique()
 
# Create a dictionary comprehension and assign it to a variable
df_DA_month = {month: df_DA[df_DA["job_posted_month"] == month] for month in months}

# Create a list of DataFrames for each month in Q1
quarter_1 = [df_DA_month["Jan"], df_DA_month["Feb"], df_DA_month["Mar"]]

# Concatenate the DataFrames in the list
df_concat = pd.concat(quarter_1, ignore_index=True)

# Display the concatenated DataFrame
df_concat 
```
