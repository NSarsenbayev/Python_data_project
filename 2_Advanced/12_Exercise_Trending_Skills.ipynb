```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df["job_title_short"] == "Data Analyst"].copy()
```

```python
df_DA["job_posted_month_no"] = df_DA["job_posted_date"].dt.month
```

```python
# df_DA_explode = df_DA["job_skills"].explode() # Only the job_skills column is exploded and returned — no other columns

df_DA_explode = df_DA.explode("job_skills") # The entire DataFrame is returned, with job_skills exploded and other columns preserved

df_DA_explode
```

```python
df_DA_explode_pivot = df_DA_explode.pivot_table(index="job_posted_month_no", columns="job_skills", 
aggfunc="size", fill_value=0)

df_DA_explode_pivot.head(3)
```

```python
# Sum across months to get total counts for each skill
df_DA_explode_pivot.loc["Total"] = df_DA_explode_pivot.sum()

df_DA_explode_pivot
```

```python
df_DA_explode_pivot.loc["Total"].sort_values(ascending=False).index

```

```python
df_DA_explode_pivot = df_DA_explode_pivot[df_DA_explode_pivot.loc["Total"].sort_values(ascending=False).index]

df_DA_explode_pivot
```

```python
df_DA_explode_pivot = df_DA_explode_pivot.drop("Total")
```

```python
df_DA_explode_pivot
```

```python
df_DA_explode_pivot.iloc[:, :5].plot()
plt.title("Top 5 skills for Data Analysts per Month")
plt.ylabel("Count")
plt.xlabel("Month")
```

```python
df_DA_explode_pivot = df_DA_explode_pivot.reset_index()
```

```python
df_DA_explode_pivot["job_posted_month"] = df_DA_explode_pivot["job_posted_month_no"].apply(lambda x: pd.to_datetime(x, format="%m").strftime("%b"))
df_DA_explode_pivot = df_DA_explode_pivot.set_index("job_posted_month")
df_DA_explode_pivot = df_DA_explode_pivot.drop(columns="job_posted_month_no")
df_DA_explode_pivot
```

```python
df_DA_explode_pivot.iloc[:, :5].plot()
plt.title("Top 5 skills for Data Analysts per Month")
plt.ylabel("Count")
plt.xlabel("")
```
